```python
game_code = coins = 0
power = 1
auto_miner = 0

print("===⛏ COIN CLICKER GAME ⛏===")

while True:

    coins += auto_miner
    print(f"\nCoins: {coins}")
    print(f"Mining Power: {power}")
    print(f"Auto Miner: {auto_miner}")

    print("\n1. Mine")
    print("2. Upgrade Pickaxe (10 coins)")
    print("3. Buy Auto Miner (25 coins)")
    print("4. Exit")

    choice = input("Choose: ")

    if choice == "1":
        coins += power
        print(f"⛏ You mined {power} coins!")

    elif choice == "2":

        if coins >= 10:
            coins -= 10
            power += 1
            print("🔥 Pickaxe upgraded!")

        else:
            print("❌ Not enough coins!")

    elif choice == "3":

        if coins >= 25:
            coins -= 25
            auto_miner += 1
            print("🤖 Auto miner purchased!")

        else:
            print("❌ Not enough coins!")

    elif choice == "4":
        print("👋 Goodbye!")
        break

with open("clicker_game.py", "w") as file:
    file.write(game_code)

print("File saved!")
```
