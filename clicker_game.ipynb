```python
import json

def save_game(coins, power, auto_miner):

    data = {
        "coins": coins,
        "power": power,
        "auto_miner": auto_miner
    }

    with open("savegame.json", "w") as file:
        json.dump(data, file)

    print("💾 Game Saved!")

def load_game():

    try:
        with open("savegame.json", "r") as file:

            data = json.load(file)

            print("✅ Save Loaded!")

            return (
                data["coins"],
                data["power"],
                data["auto_miner"]
            )

    except:
        print("⚠ No save file found!")

        return 0, 1, 0

coins, power, auto_miner = load_game()

print("=== COIN CLICKER GAME ===")

while True:

    coins += auto_miner

    print(f"\nCoins: {coins}")
    print(f"Mining Power: {power}")
    print(f"Auto Miner: {auto_miner}")

    print("\n1. Mine")
    print("2. Upgrade Pickaxe (10 coins)")
    print("3. Buy Auto Miner (25 coins)")
    print("4. Save Game")
    print("5. Exit")

    choice = input("Choose: ")

    if choice == "1":

        coins += power
        print(f"⛏ You mined {power} coins!")

    elif choice == "2":

        if coins >= 10:
            coins -= 10
            power += 1
            print("🔥 Pickaxe upgraded!")

        else:
            print("❌ Not enough coins!")

    elif choice == "3":

        if coins >= 25:
            coins -= 25
            auto_miner += 1
            print("🤖 Auto miner purchased!")

        else:
            print("❌ Not enough coins!")

    elif choice == "4":

        save_game(coins, power, auto_miner)

    elif choice == "5":

        print("👋 Goodbye!")
        break
```
